```python
def func(a, b, c, d):
    res = {'s1':np.array([0]), 's2':np.array([0]), 's3':np.array([0]), 's4':np.array([0]), 't1':np.array([0]), 't2':np.array([0]), 't3':np.array([0]), 't4':np.array([0]), 'b1':np.array([0]), 'b2':np.array([0]), 'b3':np.array([0]), 'b4':np.array([0]), 'a1':np.array([0]), 'a2':np.array([0]), 'a3':np.array([0]), 'a4':np.array([0]), }
    arr = [[0, 0, 0, 0], [8, 9, 2, 9], [2, 2, 10, 10], [7, 8, 4, 5], [5, 5, 5, 3]]
    res['s1'] = [arr[a][0]]
    res['s2'] = [arr[a][1]]
    res['s3'] = [arr[a][2]]
    res['s4'] = [arr[a][3]]
    arr = [[0, 0, 0, 0], [5, 4, 5, 3], [7, 9, 2, 6], [6, 5, 5, 2], [1, 3, 10, 8]]
    res['t1'] = [arr[b][0]]
    res['t2'] = [arr[b][1]]
    res['t3'] = [arr[b][2]]
    res['t4'] = [arr[b][3]]
    arr = [[0, 0, 0, 0], [10, 8, 3, 5], [6, 5, 5, 8], [1, 4, 7, 10], [1, 5, 9, 6]]
    res['b1'] = [arr[c][0]]
    res['b2'] = [arr[c][1]]
    res['b3'] = [arr[c][2]]
    res['b4'] = [arr[c][3]]
    arr = [[0, 0, 0, 0], [9, 9, 4, 9], [7, 8, 6, 7], [2, 6, 5, 10], [4, 7, 8, 3]]
    res['a1'] = [arr[d][0]]
    res['a2'] = [arr[d][1]]
    res['a3'] = [arr[d][2]]
    res['a4'] = [arr[d][3]]
    res = pd.DataFrame(res, columns=column)
    z = model.predict(res)
    return z[0]
```
